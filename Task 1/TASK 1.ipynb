# Dropping unnecessary columns

```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

## Extraction of top 10 countries with respect to total population

```python
# Filter data for total population

#This line filters df to keep only the rows where the Series Code is 'SP.POP.TOTL'. This code likely corresponds to the total population of each country.
total_population_data = df[df['Series Code']== 'SP.POP.TOTL'] 

#Sort data based on the total population for 2022

#sorts this data based on the population in the year 2022. ascending=False means the sorting is in descending order, so countries with the highest populations for 2022 come first.
total_population_sorted = total_population_data.sort_values(by="2022",ascending = False)

#Sort the top 10 countries with the heightest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)

print("Top ten countries of total population \n ")
print(total_top_ten_countries[['Country Code']])
```

# Bar Plot

## Top ten countries of total population in year 2022 and 2016

```python
# create a bar plot

# initializes a new figure with a size of 15 inches wide by 6 inches tall. 
plt.figure(figsize=(15,6))



# This line creates a subplot in a 2x2 grid (2 rows and 2 columns), placing the plot in the first position (top-left corner).
# This means you are preparing to create a subplot that will be part of a grid layout of up to four plots.
plt.subplot(2,2,1)



sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries, palette="viridis")
# This line uses Seaborn’s barplot function to create a bar plot.
# x="2022": The x-axis shows the population data for the year 2022.
# y="Country Code": The y-axis displays the country codes.
# Specifies the color palette for the bars. viridis is a perceptually uniform colormap suitable for showing differences in data.




plt.title("Top ten countries of total population (2022)", fontsize=10) # Sets the title of the plot.
plt.xlabel("Total population", fontsize=10) # Labels the x-axis as "Total population".
plt.ylabel("Country", fontsize=10) # : Labels the y-axis as "Country".
plt.show() # displays the plot.

# Create a bar plot for 2016
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=total_top_ten_countries, palette="viridis")

plt.title("Top ten countries of total population (2016)", fontsize=10)
plt.xlabel("Total population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show()
```

## Extration of top 10 countries with heightest male population

```python
# filter data for male population 
male_population_data = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']

#sort data based on the male population for 2022
male_population_sorted = male_population_data.sort_values(by = "2022", ascending= False)

# Get the top 10 countries with the highest male population for 2022
male_top_ten_countries= male_population_sorted.head(10)

print("Top ten countries of male population ")
print(male_top_ten_countries[['Country Code']])
```

## Top ten countries with highest male and female population in 2022

```python
# Create a bar plot for 2022
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data= male_top_ten_countries, palette="coolwarm")

plt.title("Top ten countries of male population (2022)", fontsize=10)
plt.xlabel("Male population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show()

# Create a bar plot for 2016
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data= female_top_ten_countries, palette="coolwarm")

plt.title("Top ten countries of female population (2022)", fontsize=10)
plt.xlabel("Female population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show()
```

# Stacked Bar Plot

## Top 10 countries with male and female population

```python
# merge male and female population data on 'Country Code'

merge_data = pd.merge(male_population_data, female_population_data, on="Country Code", suffixes = ("_male", "_female"))
```

```python
#merged data 

#calculate the total population for each country (male + female )
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data.head()
```

```python
#sort data based on total population in descending order

sorted_data = merge_data.sort_values(by = "Total population" , ascending = False)
```

## Select the top 10 countries with the highest total population

```python
top_10_countries = sorted_data.head(10)
```

```python
plt.figure(figsize=(12, 6))

# Plot the female population
plt.bar(top_10_countries['Country Code'], top_10_countries['2022_female'], color='red', label='Female Population')

# Plot the male population, stacking it on top of the female population
plt.bar(top_10_countries['Country Code'], top_10_countries['2022_male'], bottom=top_10_countries['2022_female'], color='green', label='Male Population')

# Labels and title
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Stacked Bar Plot of Population by Gender')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

## Bottom 10 countries with male and female population(2022)

```python
# select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
plt.figure(figsize=(12, 6))

# Plot the female population
plt.bar(bottom_10_countries['Country Code'], bottom_10_countries['2022_female'], color='red', label='Female Population')

# Plot the male population, stacking it on top of the female population
plt.bar(bottom_10_countries['Country Code'], bottom_10_countries['2022_male'], bottom=bottom_10_countries['2022_female'], color='green', label='Male Population')

# Labels and title
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Stacked Bar Plot of Population by Gender')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
